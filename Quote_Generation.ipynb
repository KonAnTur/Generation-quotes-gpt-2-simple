```
import gpt_2_simple as gpt2
```

```
gpt2.download_gpt2(model_name="124M")
```

```
file_name = 'author_quotes.txt'
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='124M',
              steps=500,
              restore_from='fresh',
              run_name='run1',
              save_every=100
              )
```

```
gpt2.generate(sess, run_name='run1',
              length=100,
              truncate='\n')
```
